```python
data = xr.open_dataset('2023.nc')
data
```

```python
data = data.sel(time=slice("2023-12-01","2023-12-31"))
data = data.sel(longitude=slice(75.25,80.25),latitude=slice(31.25,26.25,1))
data = data - 273.15
```

```python
c = 0 
Agg = data.sel(time=['2023-12-01','2023-12-02','2023-12-03']).sum(dim='time')
Agg
```

```python
Threshold = 40
for x in range(1,29):
    Agg = data.sel(time=[f'2023-12-{x}',f'2023-12-{x+1}',f'2023-12-{x+2}']).sum(dim='time')
    Agg = Agg < Threshold
    print(f'Places experiencing cold waves during 2023-12-{x} to 2023-12-{x+2}')
    for y in Agg['latitude']:
        for z in Agg['longitude']:
            if(Agg.sel(latitude=y,longitude=z) == True):
                print(f'Latitude : {y.item()} Longitude : {z.item()}')
```
